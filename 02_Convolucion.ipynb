# Punto 1

Convolucionar dos señales, la primera una señal con dos fundamentales (5 y 50 Hz) y la otra variable.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive

def convolucionar(Frecuencia = 50):
  fs = 403

  t = np.arange(0,1,1/fs)
  senal1 = np.cos(2*np.pi*t* 5) + np.cos(2*np.pi*t* 50)
  senal2 = np.cos(2*np.pi*t* Frecuencia)

  tc = np.arange(0, (fs*2-1)/fs, 1/fs)
  senalc = np.convolve(senal1, senal2)

  fig = plt.figure(figsize=(12,5))

  ax1 = plt.subplot2grid((2,2),(0,0))
  ax2 = plt.subplot2grid((2,2),(0,1))
  ax3 = plt.subplot2grid((2,2),(1,0), colspan = 2)

  ax1.plot(t,senal1, linewidth = 0.7)
  ax1.set_title('Señal 5 + 50 Hz')
  ax2.plot(t,senal2, linewidth = 0.7)
  ax2.set_title(f'Señal {Frecuencia} Hz')
  ax3.plot(tc, senalc, linewidth = 0.85)
  ax3.set_title('Convolución de las Señales')

  plt.tight_layout()
  plt.show()

grafica_interactiva = interactive(convolucionar, Frecuencia = (0, 60))
grafica_interactiva
```

# Punto 2
Basado en el anterior, variar el punto de inicio de la señal y el porcentaje de la señal

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive

def convolucionar(Frecuencia = 50, Inicio = 0.4, Porcentaje = 15):
  fs = 403

  t = np.arange(0,1,1/fs)
  senal1 = np.cos(2*np.pi*t* 5) + np.cos(2*np.pi*t* 50)
  senal2 = np.cos(2*np.pi*t* Frecuencia)

  Porcentaje = np.abs(100 - Porcentaje)
  senal2 = np.concatenate((senal2,np.zeros((int(Porcentaje*len(t)/100)))))
  senal2 = senal2[len(senal2)-len(t):]

  senal2= np.concatenate((np.zeros((int(Inicio*len(t)))),senal2))
  senal2 = senal2[:len(t)]

  tc = np.arange(0, (fs*2-1)/fs, 1/fs)
  senalc = np.convolve(senal1, senal2)

  fig = plt.figure(figsize=(12,5))

  ax1 = plt.subplot2grid((2,2),(0,0))
  ax2 = plt.subplot2grid((2,2),(0,1))
  ax3 = plt.subplot2grid((2,2),(1,0), colspan = 2)

  ax1.plot(t,senal1, linewidth = 0.7)
  ax1.set_title('Señal 5 + 50 Hz')
  ax2.plot(t,senal2, linewidth = 0.7)
  ax2.set_title(f'Señal {Frecuencia} Hz')
  ax3.plot(tc, senalc, linewidth = 0.85)
  ax3.set_title('Convolución de las Señales')

  plt.tight_layout()
  plt.show()

grafica_interactiva = interactive(convolucionar, Frecuencia = (0, 60), Inicio = (0,1,0.01), Porcentaje =(0,100,5))
grafica_interactiva
```
